In natural language processing (NLP), raw text cannot be used directly by computational models. Tokenization, lemmatization, and stemming are essential preprocessing steps that convert text into a structured form suitable for analysis.
- Tokenization splits text into individual units such as words or subwords, enabling models to process sentences piece by piece.
- Stemming and lemmatization reduce words to their base forms, which helps to group variations of a word together
These steps improve computational efficiency, reduce vocabulary size, and enhance the ability
of models to generalize across similar words, making them foundational for tasks like text classification, search, and language modelling.
